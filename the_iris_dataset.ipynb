```python

#Importing the relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import pearsonr
```

```python
iris_data=sns.load_dataset('iris')
```

```python
iris_data.head()
```

```python
# 1. Scatter plot: Sepal Length vs Petal Length
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=iris_data,
    x="sepal_length",
    y="petal_length",
    hue="species",
    palette="Set2",
    s=70,
    edgecolor="k"
)
plt.title("Scatter Plot: Sepal Length vs Petal Length", fontsize=14)
plt.show()
```

The scatter plot shows a positive trend: flowers with longer sepals tend to have longer petals.


```python
# 2. Joint distribution (with kernel density estimate)
sns.jointplot(
    data=iris_data,
    x="sepal_length",
    y="petal_length",
    kind="kde",
    fill=True,
    cmap="viridis"
)
plt.suptitle("Joint Distribution of Sepal Length and Petal Length", y=1.02, fontsize=14)
plt.show()
```

The joint distribution (density plot) highlights clusters of high density that align with flower species:

Setosa occupies the bottom-left corner (**short sepal, short petal).**

Versicolor is in the middle range.

Virginica lies in the upper-right (long sepal, long petal).

```python
# 3. Correlation analysis
corr, p_value = pearsonr(iris_data["sepal_length"], iris_data["petal_length"])
print(f"Pearson correlation: {corr:.2f}")
print(f"P-value: {p_value:.4f}")
```

The Pearson correlation coefficient = 0.87, with a p-value < 0.001.

This means there is a very strong, statistically significant positive linear relationship between sepal length and petal length.

Practically, when sepal length increases, petal length also tends to increase almost proportionally.
